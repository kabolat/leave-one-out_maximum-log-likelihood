# PSCC'24 - Experiments - Tabulation and Visualization of the Results

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set()
from matplotlib.ticker import FormatStrFormatter
```

## Read Dataframes

```python
base_df_europe = pd.read_pickle('outputs/base_df_europe.pkl')
exp_df_europe = pd.read_pickle('outputs/exp_df_europe.pkl')
base_df_denmark = pd.read_pickle('outputs/base_df_denmark.pkl')
exp_df_denmark = pd.read_pickle('outputs/exp_df_denmark.pkl')
```

## ECDFs of Scores

```python
label_list = ['A-KDE',r'$\pi$-KDE',r'GMM$_A$',r'GMM$_\pi$']
```

```python
exp_df_dataset = exp_df_europe
base_df_dataset = base_df_europe
fig, axs = plt.subplots(1,2,figsize=(10,4))

scores_array = np.ndarray((len(exp_df_dataset['scores'].columns), len(exp_df_dataset['scores']), len(exp_df_dataset['scores'].iloc[0,0])))

for j, col in enumerate(exp_df_dataset['scores'].columns):
    for i in range(len(exp_df_dataset['scores'])):
        sns.ecdfplot(data=np.array(exp_df_dataset['scores'][col].iloc[i]), label=label_list[i],  ax=axs[j], lw=2, alpha=0.8)
        scores_array[j,i,:] = np.array(exp_df_dataset['scores'][col].iloc[i])
    sns.ecdfplot(data=np.array(base_df_dataset['scores'][col][0]), label="Train set", ax=axs[j], lw=2,)
    scores_array[j,-1,:] = np.array(base_df_dataset['scores'][col][0])
axs[1].set_ylabel('')
axs[0].set_xlabel('Sample Test Score (MMD)')
axs[1].set_xlabel('Sample Test Score (Energy)')
axs[1].set_yticklabels([])
axs[0].set_xlim([np.min(np.percentile(scores_array[0],q=1.0,axis=-1)),np.max(np.percentile(scores_array[0],q=99,axis=-1))])
axs[1].set_xlim([np.min(np.percentile(scores_array[1],q=1.0,axis=-1)),np.max(np.percentile(scores_array[1],q=99,axis=-1))])
plt.legend(loc='lower right')
plt.savefig("outputs/scores_europe.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
